```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('../output.csv')
```

```python
data.head()

ft = ['danceability','energy','loudness','valence','tempo']
```

```python
scaler = MinMaxScaler()

scaled = pd.DataFrame(scaler.fit_transform(data[ft]))
```

```python
scaled.columns = ft
```

```python
scaled
```

```python
kmeans = KMeans(n_clusters=4).fit(scaled)
y_kmeans = kmeans.predict(scaled)
centroids = kmeans.cluster_centers_
```

```python
y_kmeans
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
```

```python
df_pca = pca.fit_transform(scaled)
```

```python
df_pca
```

```python
prin_df = pd.DataFrame(data=df_pca, columns = ['pC1','pC2','pC3'])
```

```python
prin_df['cluster'] = y_kmeans
```

```python
prin_df
```

```python
scaled['cluster'] = y_kmeans
```

```python
import plotly.express as px

fig = px.scatter_3d(scaled, x='valence', y='energy', z='danceability', color='cluster', size_max=10)

fig.show()
```
