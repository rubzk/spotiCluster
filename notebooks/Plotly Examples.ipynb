```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('../output.csv')
```

```python
scaler = MinMaxScaler()
std = StandardScaler()
features = ['danceability', 'energy', 'loudness', 'speechiness','acousticness','instrumentalness','liveness','valence','tempo']
data[features] = scaler.fit_transform(data[features])


```

```python
audio_ft = features
features = ['danceability','energy','valence','tempo']
```

```python
kmeans = KMeans(n_clusters=4).fit(data[features])
y_kmeans = kmeans.predict(data[features])
centroids = kmeans.cluster_centers_
```

```python
cluster_stats = pd.DataFrame(columns=audio_ft)

for n in data['cluster'].value_counts().index.to_list():   
    cluster_stats = cluster_stats.append(data[data.cluster == n][audio_ft].mean(), ignore_index=True)
```

```python
cluster_stats['cluster'] = cluster_stats.index
```

```python
data[audio_ft]
```

```python
cluster_stats.loc[0]['danceability']
```

```python
import plotly.graph_objects as go
categories = cluster_stats.loc[:, cluster_stats.columns != 'cluster'].columns

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[cluster_stats.loc[0]['danceability'], cluster_stats.loc[0]['energy'], cluster_stats.loc[0]['loudness'],
         cluster_stats.loc[0]['speechiness'], cluster_stats.loc[0]['acousticness'], cluster_stats.loc[0]['instrumentalness'],
        cluster_stats.loc[0]['liveness'], cluster_stats.loc[0]['valence'], cluster_stats.loc[0]['tempo']],
      theta=categories,
      fill='toself',
      name='Cluster 0'
))
fig.add_trace(go.Scatterpolar(
      r=[cluster_stats.loc[1]['danceability'], cluster_stats.loc[1]['energy'], cluster_stats.loc[1]['loudness'],
         cluster_stats.loc[1]['speechiness'], cluster_stats.loc[1]['acousticness'], cluster_stats.loc[1]['instrumentalness'],
        cluster_stats.loc[1]['liveness'], cluster_stats.loc[1]['valence'], cluster_stats.loc[1]['tempo']],
      theta=categories,
      fill='toself',
      name='Cluster 1'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
  showlegend=False
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

categories = cluster_stats.loc[:, cluster_stats.columns != 'cluster'].columns

for i in range(0,4):
    fig = px.line_polar(cluster_stats[cluster_stats.cluster==i],
                    r=[cluster_stats.loc[i]['danceability'], cluster_stats.loc[i]['energy'], cluster_stats.loc[i]['loudness'],
         cluster_stats.loc[i]['speechiness'], cluster_stats.loc[i]['acousticness'], cluster_stats.loc[i]['instrumentalness'],
        cluster_stats.loc[i]['liveness'], cluster_stats.loc[i]['valence'], cluster_stats.loc[i]['tempo']],
                    theta=categories,
                    line_close=True)
    fig.update_traces(fill='toself')
    fig.show()

```
